```python
import pandas as pd

data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Product': ['Laptop', 'Mobile', 'Laptop', 'Tablet', 'Mobile', 'Tablet', 'Laptop', 'Mobile', 'Tablet', 'Laptop'],
    'Sales_Amount': [50000, 25000, 52000, 18000, 27000, 19000, 51000, 26000, 20000, 55000],
    'Quantity': [5, 10, 4, 6, 9, 7, 5, 11, 8, 5]
}

df = pd.DataFrame(data)
print(df)

# East region sales
east_sales = df[df["Region"] == "East"]
print(east_sales)

# Sort by Sales Amount
sorted_df = df.sort_values("Sales_Amount")
print(sorted_df)

# Region-wise total sales
region_sales = df.groupby("Region")["Sales_Amount"].sum()
print(region_sales)
```

```python
import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 65, 70, 72, 75, 80, 85, 90, 95]

plt.hist(marks, bins=12)

plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [35, 40, 50, 60, 65, 75, 85, 95]

plt.scatter(hours, marks)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```
