```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Name': ['Ramya', 'Rahima', 'Loga', 'Kavi', 'Mabitha',
             'Arun', 'Naveen', 'Charan', 'Divya', 'Elan',
             'Fathima', 'Gokul', 'Hari', 'Indhu', 'Jeeva',
             'Karthik', 'Lavanya', 'Manoj', 'Nisha', 'Sneha'],


    'Tamil':  [99,40,45,30,35, 90,92,88,91,87, 93,94,89,92,90, 95,76,88,67,78],
    'English':[97,92,70,35,30, 88,91,90,85,87, 92,93,89,94,88, 96,45,90,17,65],
    'Maths':  [100,78,62,25,20, 95,96,90,92,89, 97,98,91,95,93, 99,19,92,38,99],
    'Science':[100,55,50,30,25, 90,92,89,91,88, 94,95,90,93,89, 96,77,91,58,99],
    'Social': [98,80,68,28,30, 91,93,90,92,89, 95,96,91,94,90, 97,62,92,59,50]
}

df = pd.DataFrame(data)


df.head(20)
```

```python
df['Average'] = df[['Tamil','English','Maths','Science','Social']].mean(axis=1)

def grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 50:
        return 'C'
    else:
        return 'Fail'

df['Grade'] = df['Average'].apply(grade)
df['Result'] = df['Average'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')
df['Rank'] = df['Average'].rank(ascending=False)
print(" Student Performance Analysis")
display(df.head(20))
```

```python
topper = df.loc[df['Average'].idxmax()]
print("\n Topper:")
print(topper)
```

```python
weak_students = df[df['Result'] == 'Fail']
print("\nWeak Students:")
display(weak_students)
```

```python
plt.figure(figsize=(12,5))

# Bar Chart
plt.subplot(1,2,1)
colors = ['red','blue','green','orange','purple','cyan','magenta','yellow','pink','brown',
          'gray','lime','navy','teal','gold','coral','aqua','violet','indigo','maroon']

plt.bar(df['Name'], df['Average'], color=colors)
plt.title("Average Marks")
plt.xticks(rotation=60)
plt.show()
```

```python
# Pie Chart
plt.subplot(1,2,2)
grade_count = df['Grade'].value_counts()

plt.pie(grade_count,
        labels=grade_count.index,
        autopct='%1.1f%%',
        colors=['Gray','lightgreen','skyblue','orange'])

plt.title("Grade Distribution")


plt.tight_layout()
plt.show()
```
