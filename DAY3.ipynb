```python
def calculate_average(m1, m2, m3):
    return (m1 + m2 + m3) / 3

try:
    mark1 = float(input("Enter Mark 1: "))
    mark2 = float(input("Enter Mark 2: "))
    mark3 = float(input("Enter Mark 3: "))

    avg = calculate_average(mark1, mark2, mark3)

    print("Average =", avg)

    if avg >= 75:
        print("Grade: A")
    elif avg >= 60:
        print("Grade: B")
    elif avg >= 50:
        print("Grade: C")
    else:
        print("Grade: F")

except ValueError:
    print("Please enter valid numbers")
```
